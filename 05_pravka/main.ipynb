```python
import gymnasium as gym
import numpy as np
import time
import datetime

import import_ipynb
from dqn_p5 import DQN
from dqn_p5 import DQN_search
```

```python


game_name = "CartPole-v0"
env = gym.make(game_name)
in_states = env.observation_space.shape[0]
out_actions = env.action_space.n

env.close()

N_TESTS = 5
s_res_cnt = 0
this_time_start = time.time()

ct = datetime.datetime.now()
ct = str(ct)
ct = ct.replace(":", "-")
ct = ct.replace(" ", "_")
ct = ct[:-7]
ct = '00/'+ct
# '0_5_3_a/'

for i in range(N_TESTS):
    print(i)
    search = DQN_search(dqn_model = DQN(ct = ct+'/'+str(i), game_name = game_name, mini_batch_size = 64, num_divisions = 50))
    atypes = ['', 'ReLU', 'ReLU', 'Linear']
    search.set_a_type_array(atypes)
    search.set_lr(0.1)
    start_net_shape = np.array([in_states, 3, 3, out_actions])
    shape_steps = np.array([[0, 1, 0, 0], [0, 0, 1, 0]])
    final_shape = np.array([in_states, 10, 10, out_actions])
    n_models = 3
    TSR = 95
    test_res, shape, good_policy = search.search(n_models, TSR, epochs = 10000, 
                                                 start_shape = np.array(start_net_shape), 
                                                 shape_steps = shape_steps,
                                                 final_shape = final_shape,
                                                 adaptive_n_models = True,
                                                 min_best_result = -10
                                                )
    if (test_res!= False):
        s_res_cnt+=1

this_time_end = time.time()
diff_time = (this_time_end - this_time_start)/60
print(diff_time)
```

```python
print(100.0*s_res_cnt/N_TESTS)
```
